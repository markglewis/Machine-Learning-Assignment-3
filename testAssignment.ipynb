Import Libraries

```python
import scipy.io
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
```

Importing sklearn libraries for Machine learning.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
from sklearn.metrics import classification_report, accuracy_score
```

```python
# loading the matlab file
data = scipy.io.loadmat("assignment.mat");

#Converting to .csv file
for i in data:
    if '__' not in i:
        print(i)
        np.savetxt((i  + '.csv'), csr_matrix(data[i], dtype=np.int8).toarray(), fmt="%s", delimiter=',')
```

```python
trainingFeatures = pd.read_csv('training.csv', header=None);
```

```python
testFeatures = pd.read_csv('test.csv', header=None)
```

```python
trainingClassifier = pd.read_csv('ctrain.csv', header=None)
```

```python
trainingFeatures.head()
```

```python
testFeatures.head()
```

```python
#Training Features
features = trainingFeatures

#Training Classifier
classifier = trainingClassifier

#Test Features
test = testFeatures
#need to predict the classifier for test
```

```python
#Training the model using logistic regression

logmodel = LogisticRegression()
logmodel.fit(features, classifier)
```

## Testing the Model
Now its time to perdict the 'Classifications' our our Test Model, given the Test Features

```python
predict = logmodel.predict(test)
print(predict)
```

```python
# Write 'perdict' to a csv file
df = pd.DataFrame(
{
    'class': predict
})
df.index = df.index + 1
df.to_csv('submit.csv')
```
